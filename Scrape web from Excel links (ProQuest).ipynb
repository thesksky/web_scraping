```python
import requests as r
from bs4 import BeautifulSoup as bs

import pandas as pd
```

```python
df = pd.read_excel (r'C:\Users\thesk\Desktop\ProQuest.xlsx')
df.head(2)
```

```python
urls = list(df['DocumentURL'])
```

```python
universities = []

for url in urls:
    '''
    read each url in a dataframe column, apply requests for scraping, 
    apply Beautiful Soup to be able to use bs attributes for parsing,
    extract university which in this case is in <span class="titleAuthorETC ">
    '''
    result = r.get(url)    
    soup = bs(result.text, 'lxml')
    for aun_tag in soup.find_all(attrs={'class' : 'titleAuthorETC'}):
        aun_tag.text
    universities.append(aun_tag.text)
    
df['universities'] = universities
df.head(2)
```

```python
soup.find("article").div
```

```python
headers = []

for url in urls:
    result = r.get(url)    
    soup = bs(result.text, 'lxml')
    for h2_tag in soup.find_all('h2'):
        h2_tag.text
    headers.append(h2_tag.text)
    
df['titles'] = headers
```

```python
# Just saving to Excel gives warning that it exceeds Excel's limit for URLS
# Create an ExcelWriter object with the option not to convert strings to URLS
writer = pd.ExcelWriter(r'C:\Users\thesk\Desktop\exported_ProQuest.xlsx', engine='xlsxwriter',options={'strings_to_urls': False})
df.to_excel(writer)
writer.close()
```
